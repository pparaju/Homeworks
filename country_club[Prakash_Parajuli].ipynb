 **QUESTION7:** Produce a list of all members who have used a tennis court.
Include in your output the name of the court, and the name of the member
formatted as a single column. Ensure no duplicate data, and order by
the member name.

```python
df7 = pd.read_sql_query('''SELECT DISTINCT CONCAT(m.surname,' ',m.firstname) AS Member, f.name AS court_name
FROM Members m
INNER JOIN Bookings b 
    ON m.memid = b.memid
INNER JOIN Facilities f 
    ON b.facid = f.facid
WHERE f.name LIKE 'Tennis Court%'
ORDER BY Member;''', engine)
df7
```

**QUESTION8:** Produce a list of bookings on the day of 2012-09-14 which
will cost the member (or guest) more than $30. Remember that guests have
different costs to members (the listed costs are per half-hour 'slot'), and
the guest user's ID is always 0. Include in your output the name of the
facility, the name of the member formatted as a single column, and the cost.
Order by descending cost, and do not use any subqueries.

```python
df8 = pd.read_sql_query('''SELECT 
    f.name AS facility_name, 
    CONCAT(m.surname,' ',m.firstname) AS Member,
    CASE 
        WHEN b.memid = 0 THEN b.slots * f.guestcost
        ELSE b.slots * f.membercost 
    END AS total_cost
FROM 
    Bookings AS b
LEFT JOIN 
    Facilities AS f ON b.facid = f.facid
LEFT JOIN 
    Members AS m ON b.memid = m.memid
WHERE 
    DATE(b.starttime) = '2012-09-14' 
    AND (
        (b.memid = 0 AND b.slots * f.guestcost > 30) 
        OR (b.memid <> 0 AND b.slots * f.membercost > 30)
    )
ORDER BY 
    total_cost DESC;''', engine)
df8
```

**QUESTION9:** This time, produce the same result as in Q8, but using a subquery.

```python
df9 = pd.read_sql_query('''SELECT 
    facility_name, 
    m.surname, 
    m.firstname,
    total_cost
FROM (
    SELECT 
        b.facid,
        b.memid,
    	f.name AS facility_name,
        CASE 
            WHEN b.memid = 0 THEN b.slots * f.guestcost
            ELSE b.slots * f.membercost 
        END AS total_cost
    FROM 
        Bookings AS b
    LEFT JOIN 
        Facilities AS f 
    	ON b.facid = f.facid
    WHERE 
        DATE(b.starttime) = '2012-09-14'
) AS totalcost
LEFT JOIN 
    Members AS m ON totalcost.memid = m.memid
WHERE total_cost > 30
ORDER BY 
    total_cost DESC;''', engine)
df9
```

**QUESTION10:** Produce a list of facilities with a total revenue less than 1000.
The output of facility name and total revenue, sorted by revenue. Remember
that there's a different cost for guests and members!

```python
df10 = pd.read_sql_query('''SELECT 
    facility_name,
    SUM(revenue) AS total_revenue
FROM (
    SELECT 
        f.name AS facility_name,
        CASE 
            WHEN b.memid = 0 THEN b.slots * f.guestcost
            ELSE b.slots * f.membercost 
        END AS revenue
    FROM 
        Bookings AS b
    LEFT JOIN 
        Facilities AS f ON b.facid = f.facid
) AS revenue_per_facility
GROUP BY 
    facility_name
HAVING 
    SUM(revenue) < 1000
ORDER BY total_revenue;''', engine)
df10
```

**QUESTION11:** Produce a report of members and who recommended them in alphabetic surname,firstname order

```python
df11 = pd.read_sql_query('''SELECT CONCAT(m1.surname,' ',m1.firstname) AS Member,
       CONCAT(m2.surname,' ',m2.firstname) AS Recommendor
FROM Members AS m1
INNER JOIN Members AS m2 
ON m1.recommendedby = m2.memid
ORDER BY Member,Recommendor;''', engine)
df11
```

**QUESTION12:** Find the facilities with their usage by member, but not guests

```python
df12 = pd.read_sql_query('''SELECT f.name AS Facility,
       SUM(b.slots) AS total_slots,
       COUNT(DISTINCT b.memid) AS used_by,
       ROUND(SUM(b.slots) / COUNT(DISTINCT b.memid), 2) AS avg_per_person
FROM Bookings AS b
LEFT JOIN Facilities AS f ON b.facid = f.facid
WHERE b.memid <> 0
GROUP BY f.name
ORDER BY total_slots DESC;''', engine)
df12
```

**QUESTION13:** Find the facilities usage by month, but not guests

```python
df13 = pd.read_sql_query('''
SELECT f.name AS Facility,
       SUM(b.slots) AS total_slots,
       COUNT(DISTINCT b.memid) AS used_by,
       ROUND(SUM(b.slots) / COUNT(DISTINCT b.memid), 2) AS avg_per_person,
       COUNT(DISTINCT strftime('%m', b.starttime)) AS months_used
FROM Bookings AS b
LEFT JOIN Facilities AS f ON b.facid = f.facid
WHERE b.memid <> 0
GROUP BY f.name
ORDER BY total_slots DESC;
''', engine)
df13
```
