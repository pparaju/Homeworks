In the United States, animal bites are often reported to law enforcement (such as animal control). The main concern with an animal bite is that the animal may be rabid. This dataset includes information on over 9,000 animal bites which occurred near Louisville, Kentucky from 1985 to 2017 and includes information on whether the animal was quarantined after the bite occurred and whether that animal was rabid.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import missingno as msno
import seaborn as sns
```

```python
Bites_df = pd.read_csv("Health_AnimalBites.csv")
```

```python
Bites_df.head()
```

Attributes of animal bite incidents reported to and investigated by Louisville Metro Department of Public Health and Wellness. Personal/identifying data has been removed. This dataset is a single .csv with the following fields.

bite_date: The date the bite occurred

SpeciesIDDesc: The species of animal that did the biting

BreedIDDesc: Breed (if known)

GenderIDDesc: Gender (of the animal)

color: color of the animal

vaccination_yrs: how many years had passed since the last vaccination

vaccination_date: the date of the last vaccination

victim_zip: the zipcode of the victim

AdvIssuedYNDesc: whether advice was issued

WhereBittenIDDesc: Where on the body the victim was bitten

quarantine_date: whether the animal was quarantined

DispositionIDDesc: whether the animal was released from quarantine

head_sent_date: the date the animalâ€™s head was sent to the lab

release_date: the date the animal was released

ResultsIDDesc: results from lab tests (for rabies)

```python
# List of columns to convert to datetime
date_columns = ['bite_date', 'vaccination_date', 'quarantine_date', 'head_sent_date', 'release_date']

# Convert the listed columns to datetime
Bites_df[date_columns] = Bites_df[date_columns].apply(pd.to_datetime, errors='coerce')
```

```python
Bites_df.describe()
```

```python
start_date = pd.Timestamp('1985-01-01')
end_date = pd.Timestamp('2017-12-31')

Bites_df.loc[(Bites_df['bite_date'] < start_date) | (Bites_df['bite_date'] > end_date), 'bite_date'] = np.nan
Bites_df.loc[(Bites_df['vaccination_date'] > end_date), 'vaccination_date'] = np.nan
Bites_df.loc[(Bites_df['release_date'] > end_date), 'release_date'] = np.nan
```

```python
Bites_df = Bites_df.sort_values(by='bite_date').reset_index(drop=True)
```

```python
Bites_df.describe()
```

```python
Bites_df.head()
```

```python
sns.set()
Bites_df['year'] = Bites_df['bite_date'].dt.year
# Group by year and count occurrences
year_counts = Bites_df.groupby('year').size()


norm = plt.Normalize(min(year_counts.values), max(year_counts.values))

cmap = plt.get_cmap('coolwarm')

# Get colors based on the normalized values
colors = cmap(norm(year_counts.values))
# Plot bar graph for number of entries by year
plt.figure(figsize=(12, 6))
plt.bar(year_counts.index.astype(str), year_counts.values, color=colors, edgecolor='black')

# Customize tick labels
plt.title('Number of bites by Year')
plt.xlabel('Year')
plt.ylabel('Number of bites')
plt.xticks(rotation=45)  # Rotate labels for readability

plt.tight_layout()
plt.savefig('bites_distribution_by_year.png', format='png', dpi=300)
plt.show()
```

```python
Bites_df['month'] = Bites_df['bite_date'].dt.month_name()
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
                'July', 'August', 'September', 'October', 'November', 'December']
Bites_df['month'] = pd.Categorical(Bites_df['month'], categories=month_order, ordered=True)

month_counts = Bites_df['month'].value_counts().reindex(month_order)
month_counts
```

```python

# Normalize the values to map them to a color scale
norm = plt.Normalize(min(month_counts.values), max(month_counts.values))

cmap = plt.get_cmap('coolwarm')

# Get colors based on the normalized values
colors = cmap(norm(month_counts.values))

# Create the plot and specify the axes
fig, ax1 = plt.subplots(figsize=(10, 5))

# Plot the bar graph with colored bars on the primary y-axis
bars = ax1.bar(month_counts.index, month_counts.values, color=colors)

# Customize the primary y-axis
ax1.set_title('Distribution of Bites by Month')
ax1.set_xlabel('Month of the Year')
ax1.set_ylabel('Number of Bites')
ax1.set_xticks(month_counts.index)
ax1.set_xticklabels(month_counts.index, rotation=90)

plt.tight_layout()
plt.savefig('bites_distribution_by_month.png', format='png', dpi=300)
plt.show()
```

```python
# Extract month while handling NaT (Not a Time) values
Bites_df['day_of_week'] = Bites_df['bite_date'].dt.day_name()
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
Bites_df['day_of_week'] = pd.Categorical(Bites_df['day_of_week'], categories=days_order, ordered=True)

# Group by day_of_week and count occurrences
day_counts = Bites_df['day_of_week'].value_counts().reindex(days_order)
day_counts
```

```python
# Create the plot and specify the axes
fig, ax1 = plt.subplots(figsize=(10, 5))

# Normalize the values to map them to a color scale
norm = plt.Normalize(min(day_counts.values), max(day_counts.values))

cmap = plt.get_cmap('coolwarm')

# Get colors based on the normalized values
colors = cmap(norm(day_counts.values))

# Plot the bar graph with colored bars on the primary y-axis
bars = ax1.bar(day_counts.index, day_counts.values, color=colors)

# Customize the primary y-axis
ax1.set_title('Distribution of Bites by Day of the Week')
ax1.set_xlabel('Day of the Week')
ax1.set_ylabel('Number of Bites')
ax1.set_xticks(day_counts.index)
ax1.set_xticklabels(day_counts.index)

plt.tight_layout()
plt.savefig('bites_distribution_by_day_of_week.png', format='png', dpi=300)
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(6, 5))

# Normalize the values to map them to a color scale
norm = plt.Normalize(min(species_counts.values), max(species_counts.values))
cmap = plt.get_cmap('coolwarm')

# Get colors based on the normalized values
colors = cmap(norm(species_counts.values))

# Plot the bar graph with colored bars on the primary y-axis
bars = ax1.bar(species_counts.index, species_counts.values, color=colors)

# Annotate the values on top of each bar
for bar in bars:
    yval = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2, yval, int(yval), ha='center', va='bottom', fontsize=10)

# Customize the primary y-axis
ax1.set_title('Distribution of Bites by Species')
ax1.set_xlabel('Species type')
ax1.set_ylabel('Number of Bites')
ax1.set_xticks(species_counts.index)
ax1.set_xticklabels(species_counts.index)

plt.tight_layout()
plt.savefig('bites_distribution_by_species.png', format='png', dpi=300)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Get the value counts and identify the top 10 breeds
breed_counts = Bites_df['BreedIDDesc'].value_counts()
top_10_breeds = breed_counts[:15]
# Normalize the values to map them to a color scale
norm = plt.Normalize(min(top_10_breeds.values), max(top_10_breeds.values))
cmap = plt.get_cmap('coolwarm')

# Get colors based on the normalized values
colors = cmap(norm(top_10_breeds.values))

# Plot the histogram
top_10_breeds.plot(kind='bar', color=colors, edgecolor='black')
plt.title('Ditribution of bites by Breed (top 15)')
plt.xlabel('Breed of the animal')
plt.ylabel('Recorded Bite Counts')
plt.xticks(rotation=90)
plt.savefig('bites_distribution_by_breed.png', format='png', dpi=300)
plt.show()
```

```python
Bites_by_zip = pd.DataFrame( Bites_df['victim_zip'].value_counts()).reset_index()
Bites_by_zip = Bites_by_zip.rename(columns={"victim_zip": "zip_code"})
Bites_by_zip.head()
```

```python
population_df = pd.read_csv("population.csv")
population_df = population_df.iloc[:,:2]
# Rename the columns
population_df = population_df.rename(columns={"ZipCode": "zip_code"})
population_df['zip_code'] = population_df['zip_code'].astype(str)
# Remove commas and strip whitespace, then convert the Population column to integers
population_df['Population'] = population_df['Population'].str.replace(',', '').str.strip().astype(int)

population_df.head()
```

```python
merged_df = pd.merge(Bites_by_zip, population_df, on="zip_code", how="left")
merged_df = merged_df[merged_df["count"]>50]
merged_df.head()
```

```python
plt.figure(figsize=(10, 6))

# Create a scatter plot with a regression line
sns.regplot(x='Population', y='count', data=merged_df, 
            scatter_kws={'color': 'blue', 'alpha': 0.7}, 
            line_kws={'color': 'red'})

# Add annotations for each point
for i in range(len(merged_df)):
    plt.annotate(merged_df['zip_code'][i], 
                 (merged_df['Population'][i], merged_df['count'][i]), 
                 textcoords="offset points", 
                 xytext=(0,5), 
                 ha='center')

# Customize the plot
plt.title('Scatter Plot of Population vs. Count with Fit Line')
plt.xlabel('Population')
plt.ylabel('Count')
plt.grid(True)
plt.savefig('bite_count_population2.png', format='png', dpi=300)
plt.show()
```

```python
species_df = pd.DataFrame(Bites_df["SpeciesIDDesc"].value_counts()).reset_index().rename(columns={"SpeciesIDDesc": "Species","count":"Total"})

vaccinated = pd.DataFrame(Bites_df[(Bites_df['vaccination_yrs'] > 0)]["SpeciesIDDesc"].value_counts()).reset_index().rename(columns={"SpeciesIDDesc": "Species","count":"Vaccinated"})
species_df = pd.merge(species_df,vaccinated, on= "Species", how = "left")

unvaccinated = pd.DataFrame(Bites_df[~(Bites_df['vaccination_yrs'] > 0)]["SpeciesIDDesc"].value_counts()).reset_index().rename(columns={"SpeciesIDDesc": "Species","count":"Unvaccinated"})
species_df = pd.merge(species_df,unvaccinated, on= "Species", how = "left")

sent_for_test = pd.DataFrame(Bites_df[((Bites_df['ResultsIDDesc'] == "NEGATIVE") | (Bites_df['ResultsIDDesc'] == "POSITIVE"))]["SpeciesIDDesc"].value_counts()).reset_index().rename(columns={"SpeciesIDDesc": "Species","count":"sent_for_test"})
species_df = pd.merge(species_df,sent_for_test, on= "Species", how = "left")

vaccinated_and_tested_positive = pd.DataFrame(Bites_df[((Bites_df['vaccination_yrs'] > 0)&(Bites_df['ResultsIDDesc'] == "POSITIVE" ))]["SpeciesIDDesc"].value_counts()).reset_index().rename(columns={"SpeciesIDDesc": "Species","count":"vaccinated_and_ntested_positive"})
species_df = pd.merge(species_df,vaccinated_and_tested_positive, on= "Species", how = "left")

vaccinated_and_tested_negative = pd.DataFrame(Bites_df[((Bites_df['vaccination_yrs'] > 0)&(Bites_df['ResultsIDDesc'] == "NEGATIVE" ))]["SpeciesIDDesc"].value_counts()).reset_index().rename(columns={"SpeciesIDDesc": "Species","count":"Vaccinated_and_tested_negative"})
species_df = pd.merge(species_df,vaccinated_and_tested_negative, on= "Species", how = "left")

unvaccinated_and_tested_positive = pd.DataFrame(Bites_df[(~(Bites_df['vaccination_yrs'] > 0)& (Bites_df['ResultsIDDesc'] == "POSITIVE"))]["SpeciesIDDesc"].value_counts()).reset_index().rename(columns={"SpeciesIDDesc": "Species","count":"Unvaccinated_and_tested_positive"})
species_df = pd.merge(species_df,unvaccinated_and_tested_positive, on= "Species", how = "left")

unvaccinated_and_tested_negative = pd.DataFrame(Bites_df[(~(Bites_df['vaccination_yrs'] > 0)&(Bites_df['ResultsIDDesc'] == "NEGATIVE" ))]["SpeciesIDDesc"].value_counts()).reset_index().rename(columns={"SpeciesIDDesc": "Species","count":"Unvaccinated_but_tested_negative"})
species_df = pd.merge(species_df,unvaccinated_and_tested_negative, on= "Species", how = "left")
species_df = species_df.fillna(0)
species_df = species_df.set_index("Species")
species_df
species_df.to_excel('species_analysis.xlsx',)
```
